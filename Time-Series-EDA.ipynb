```python
# Importing Libraries

import pandas as pd
import yfinance as yf
```

```python
# Importing data

data = yf.download('aapl')
data.head()
```

```python
data.to_csv(rf'./Data/APPLE.csv')
```

```python
df = data
df['High'].plot(figsize=(10,4))
```

```python
# Xlimit and ylimit
df['High'].plot(xlim=['2015-01-01', '2023-01-01'], ylim=[0,200],figsize=(10,4))
```

```python
# Xlimit and ylimit and coloring
df['High'].plot(xlim=['2015-01-01', '2023-01-01'], ylim=[0,200],figsize=(10,4), ls='--', c='green')
```

```python
df.index
```

```python
index = df.loc['2015-01-01':'2023-01-01'].index
share_open = df.loc['2015-01-01':'2023-01-01']['Open']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure,axis = plt.subplots()
plt.tight_layout()
## Preventing overlapping
figure.autofmt_xdate()
axis.plot(index, share_open)
```

# Time Resampling

```python
df.head()
```

```python
df.resample(rule='A').min()
```

```python
df.resample(rule='A').min()['Open'].plot()
```

```python
df.resample(rule='A').min()['High'].plot()
```

```python
df.resample(rule='BA').min()
```

```python
df['Open: 50 days rolling'] = df['Open'].rolling(50).mean()
```

```python
df.head(60)
```

```python
df[['Open', 'Open: 50 days rolling']].plot(figsize=(12,5))
```
